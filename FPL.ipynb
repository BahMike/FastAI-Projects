```python
from fastai import *
from fastai.tabular import *
```

```python
df = pd.read_csv('playerdatav2.csv')
```

```python
df.head()
```

```python
df = df.sort_values(by='kickoff_time', ascending=False)
```

```python
add_datepart(df, "kickoff_time", drop=False)
```

```python
def display_all(df):
    with pd.option_context("display.max_rows", 1000, "display.max_columns", 1000): 
        display(df)
        
display_all(df.tail(10).T)
```

```python
# Defining pre-processing we want for our fast.ai DataBunch
procs=[Categorify, Normalize]
```

```python
df.dtypes
g = df.columns.to_series().groupby(df.dtypes).groups
g
```

```python
cat_names = ['was_home', 'kickoff_timeIs_month_end', 'kickoff_timeIs_month_start',
        'kickoff_timeIs_quarter_end', 'kickoff_timeIs_quarter_start',
        'kickoff_timeIs_year_end', 'kickoff_timeIs_year_start', 'element']
```

```python
cont_vars = list(set(df)-set(cat_names)-{dep_var})
```

```python
dep_var = 'transfers_balance'
df2 = df[cat_names + cont_vars + [dep_var,'kickoff_time']].copy()
```

```python
# Look at the time period of test set, make sure it's more recent
df['kickoff_time'].min(), df['kickoff_time'].max()
# Calculate where we should cut the validation set. We pick the most recent 'n' records in training set where n is the number of entries in test set. 
cut = df['kickoff_time'][(df['kickoff_time'] == df['kickoff_time'][len(df)-1])].index.max()
cut
# specify the valid_idx variable as the cut out range.
valid_idx = range(cut)
```

```python
# Use fast.ai datablock api to put our training data into the DataBunch, getting ready for training
data = (TabularList.from_df(df, path='/', cat_names=cat_names, cont_names=cont_vars, procs=procs)
                   .split_by_idx(valid_idx)
                   .label_from_df(cols=dep_var, label_cls=FloatList)
                   .databunch())
```
